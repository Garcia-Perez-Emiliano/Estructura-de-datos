```python
class Pila:
    def __init__(self):
        self.items=[]

    def apilar(self,x):
        self.items.append(x)

    def despilar(self):
        try:
            return self.items.pop()
        except IndexError:
            raise ValueError("La pila esta vacia")

    def es_vacia(self):
        return self.items==[]

    def pilaimprimir(self):
        print("*********")
        reversed_items = (reversed(self.items))
        for item in reversed_items:
            print(str(item))


p1=Pila()
for i in range(3,0,-1):
    p1.apilar(i)
p2=Pila()
p3=Pila()
class Torres:

    def __init__(self):
        self.t1 = p1
        self.t2 = p2
        self.t3 = p3

    def mover(self):
        if len(self.t3.items)==3:
            print("")
        elif self.t1.es_vacia()==False:
            if self.t3.es_vacia()==True:
                self.t3.apilar(self.t1.despilar())
                self.imprimir()
                self.mover()
            elif self.t1.items[len(self.t1.items) - 1] < self.t3.items[len(self.t3.items) - 1]:
                self.t3.apilar(self.t1.despilar())
                self.imprimir()
                self.mover()
            elif self.t2.es_vacia()==True:
                self.t2.apilar(self.t1.despilar())
                self.imprimir()
                self.mover()
            elif self.t1.items[len(self.t1.items) - 1] < self.t2.items[len(self.t2.items) - 1]:
                self.t2.apilar(self.t1.despilar())
                self.imprimir()
                self.mover()
            elif self.t3.items[len(self.t3.items) - 1] < self.t2.items[len(self.t2.items) - 1]:
                self.t2.apilar(self.t3.despilar())
                self.imprimir()
                self.mover()
        elif self.t1.es_vacia()==True:
            if self.t1.es_vacia()==True:
                self.t1.apilar(self.t2.despilar())
                self.imprimir()
                if self.t2.items[len(self.t2.items) - 1] < self.t3.items[len(self.t3.items) - 1]:
                    self.t3.apilar(self.t2.despilar())
                    self.imprimir()
                    self.mover()


    def imprimir(self):
        print("\n")
        print("Torre 1")
        self.t1.pilaimprimir()
        print("Torre 2")
        self.t2.pilaimprimir()
        print("Torre 3")
        self.t3.pilaimprimir()

a=Torres()
a.imprimir()
a.mover()
```
