```python
comidas=[]
bebidas=[]
alumnos=[]

def captura():
    val=input("->")
    return val

def menu ():
    op=0
    while op!=11:
        print("1. Ingresar alumno a la fila")
        print("2. Preparar comida a la fila")
        print("3. Preparar bebidas a la fila")
        print("4. Agregar comida al inicio de la fila")
        print("5. Agregar bebida al inicio de la fila")
        print("6. Servir comida y bebida al alumno")
        print("7. Buscar si existe un nombre de alumno en la fila")
        print("8. Buscar si existe comida en la fila y cuantas veces existe")
        print("9. Buscar si existe bebida en la fila y cuantas veces existe")
        print("10. Ver alumnos, bebidad y comidas en fila")
        print("11. Salir")
        op = int(captura())
        if op==1:
            print("Numero de alumnos a ingresar")
            n = int(captura())
            ingrealumno(n)
        elif op==2:
            print("Numero de comidas a preparar")
            n = int(captura())
            ingrecomida(n)
        elif op==3:
            print("Numero de bebidas a preparar")
            n = int(captura())
            ingrebebida(n)
        elif op==4:
            inicomida()
        elif op==5:
            inibebida()
        elif op==6:
            servir()
        elif op==7:
            balumno()
        elif op==8:
            bcomida()
        elif op==9:
            bbebida()
        elif op==10:
            imprimir()

def ingrealumno(n):
    print("Ingresa el nombre del alumno")
    respuesta = captura()
    if n>0:
        alumnos.append(respuesta)
        n-=1
        ingrealumno(n)
    else:
        print("Todos los datos se han guardado")

def ingrecomida(n):
    print("Nombre de la comida preparada")
    respuesta = captura()
    if n>0:
        comidas.append(respuesta)
        n-=1
        ingrecomida(n)
    else:
        print("Todos los datos se han guardado")

def ingrebebida(n):
    print("Nombre de la bebida preparada")
    respuesta = captura()
    if n>0:
        bebidas.append(respuesta)
        n-=1
        ingrebebida(n)
    else:
        print("Todos los datos se han guardado")

def inicomida():
    print("Nombre de la comida preparada")
    respuesta = captura()
    comidas.insert(0,respuesta)

def inibebida():
    print("Nombre de la bebida preparada")
    respuesta = captura()
    bebidas.insert(0,respuesta)

def servir():
    if len(alumnos)>=1 and len(comidas)>=1 and len(bebidas)>=1:
        print("Al alumno {0} se le sirvio {1} de comida y {2} de bebida".format(alumnos[0],bebidas[0],comidas[0]))
        comidas.reverse()
        bebidas.reverse()
        alumnos.reverse()
        comidas.pop()
        bebidas.pop()
        alumnos.pop()
        comidas.reverse()
        bebidas.reverse()
        alumnos.reverse()
    else:
        print("***ERROR***")
        print("Hay",len(alumnos),"alumnos en fila")
        print("Hay", len(comidas), "comidas en fila")
        print("Hay", len(bebidas), "bebidas en fila")

def balumno():
    print("Ingrese el nombre del alumno")
    respuesta=captura()
    print(f" {respuesta} esta {alumnos.count(respuesta)}")

def bcomida():
    print("Ingrese el nombre de la comida")
    respuesta=captura()
    print(f" {respuesta} esta {comidas.count(respuesta)}")

def bbebida():
    print("Ingrese el nombre de la bebida")
    respuesta=captura()
    print(f" {respuesta} esta {bebidas.count(respuesta)}")

def imprimir():
    print("1. Orden original")
    print("2. Orden alfabetico")
    r=int(captura())
    if r==1:
        print(alumnos)
        print(comidas)
        print(bebidas)
    elif r==2:
        print(alumnos.sort())
        print(comidas.sort())
        print(bebidas.sort())
    else:
        print("Opcion invalida")
```
